```python
import requests
import _thread
from bs4 import BeautifulSoup
import time
from selenium import webdriver
```

```python
browser = webdriver.Chrome(executable_path='C:/Users/JoeZhong/.spyder-py3/chromedriver')
browser.get("https://hahow.in/courses")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
num_page = soup.find_all(class_="rc-pagination-item")[-1].text
```

```python
urls = ["https://hahow.in/courses?page={}".format(i) for i in range(1, int(num_page)+1)]
```

```python
def get_course(url, courses):
    browser = webdriver.Chrome(executable_path='C:/Users/JoeZhong/.spyder-py3/chromedriver')
    browser.get(url)
    time.sleep(1)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser')
    for h in soup.find_all("h4", class_="title"):
        courses.append(h.text)
```

```python
t1 = time.time()
courses = []
for url in urls:
    get_course(url, courses)
t2 = time.time()
print(t2-t1)
```

```python
courses
```

```python
t1 = time.time()
courses = []
for url in urls:
    _thread.start_new_thread(get_course, (url,courses))
t2 = time.time()
print(t2-t1)
```
