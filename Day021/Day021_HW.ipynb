## 1. 取出今天所有的新聞

```python
import requests
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import datetime
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
browsers = webdriver.Chrome(executable_path="C:/Users/JoeZhong/.spyder-py3/chromedriver")
browsers.get(url)
height = 0
while True:
    newheight = browsers.execute_script("return document.body.scrollHeight;")
    if height < newheight:
        browsers.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        height = newheight
    else:
        break
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime
text = browsers.page_source
soup = BeautifulSoup(text)
h3 = soup.find("div", class_="part_list_2").find_all("h3")
news = list()
for i in h3:
    dic = {}
    dic["title"] = i.a.text
    dic["type"] = i.em.text
    dic["time"] = i.span.text
    news.append(dic)
for news in news:
    time_post = datetime.datetime.strptime(news["time"], "%Y/%m/%d %H:%M")
    time_now = datetime.datetime.now()
    timedelta = datetime.timedelta(hours=2)
    if time_now-timedelta < time_post:
        print("%s %s %s" %(news["time"], news["type"], news["title"]))
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
time_now = datetime.datetime.today()
timedelta = datetime.timedelta(days=3)
time_two = time_now-timedelta
url = "https://www.ettoday.net/news/news-list-{}-0.htm".format(time_two.date())
browsers = webdriver.Chrome(executable_path="C:/Users/JoeZhong/.spyder-py3/chromedriver")
browsers.get(url)
height = 0
while True:
    newheight = browsers.execute_script("return document.body.scrollHeight;")
    if height < newheight:
        browsers.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        height = newheight
    else:
        break
```

```python
text = browsers.page_source
soup = BeautifulSoup(text)
h3 = soup.find("div", class_="part_list_2").find_all("h3")
news = list()
for i in h3:
    dic = {}
    dic["title"] = i.a.text
    dic["type"] = i.em.text
    dic["time"] = i.span.text
    news.append(dic)
for news in news:
    time_post = datetime.datetime.strptime(news["time"], "%Y/%m/%d %H:%M")
    time_three = datetime.datetime.strptime("{} 15:00".format(time_post.date()), "%Y-%m-%d %H:%M")
    timedelta = datetime.timedelta(hours=2)
    if time_three+timedelta >= time_post and time_post >= time_three:
        print("%s %s %s" %(news["time"], news["type"], news["title"]))
```
