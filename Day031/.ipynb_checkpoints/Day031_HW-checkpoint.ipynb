### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests
r = requests.get("https://goodinfo.tw/StockInfo/index.asp")
r.encoding = 'utf-8'
r.text
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
headers = {
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7',
    'cache-control': 'max-age=0',
    'cookie': 'CLIENT%5FID=20200513124352964%5F118%2E163%2E84%2E211; _ga=GA1.2.1097712021.1589345239; __gads=ID=6d8bc978cbfc319a:T=1589345236:S=ALNI_MbBwGgT2dbafv3qrPklvDl3T7aDGQ; GOOD%5FINFO%5FSTOCK%5FBROWSE%5FLIST=1%7C3044; SCREEN_SIZE=WIDTH=1536&HEIGHT=864; _gid=GA1.2.1576523485.1597312315',    'referer': 'https://www.google.com/',
    'sec-fetch-dest': 'document',
    'sec-fetch-mode': 'navigate',
    'sec-fetch-site': 'cross-site',
    'sec-fetch-user': '?1',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36'
}
r1 = requests.get("https://goodinfo.tw/StockInfo/index.asp", headers=headers)
r1.encoding = 'utf-8'
r1.text
```
