### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
bs = BeautifulSoup(r.text)
news = bs.find(class_="part_list_2").find_all("h3")
for i in news:
    print(i.span.text + " " + i.a.text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
dic_news = {i.em.text: [] for i in news}
for i in news:
    dic_news[i.em.text].append(i.a.text)
dic_news
```

### ③ 哪一個類別的文章最多

```python
max(dic_news, key=lambda x: len(dic_news[x]))
```
